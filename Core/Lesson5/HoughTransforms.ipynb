# Hough Lines
### Read in the images

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/phone.jpg').copy()
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20, 10))
plt.imshow(image)
```

### Edge detection

```python
# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Edge detection
edges = cv2.Canny(gray, 80, 100)

plt.figure(figsize=(20, 10))
plt.imshow(edges, cmap='gray')
```

### Hough transform

```python
# Hough parameters
rho = 1
theta = np.pi/180
threshold = 60
min_line_length = 100
max_line_gap = 5

# Transform
lined_image = np.copy(image)
lines = cv2.HoughLinesP(edges, rho, theta, threshold, 
                        minLineLength=min_line_length, 
                        maxLineGap=max_line_gap)

# Add lines
for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(lined_image, (x1, y1), (x2, y2), (200, 0, 255), 3)
        
plt.figure(figsize=(20, 10))
plt.imshow(lined_image)
```
