### Import resources

```python
import numpy as np
import matplotlib.image as mpimg  # for reading in images

import matplotlib.pyplot as plt
import cv2 # computer vision library

%matplotlib inline
```

### Read in and display the image

```python
# Read in the image
image = mpimg.imread('images/waymo_car.jpg')

# Print out the image dimensions
print('Image dimensions:', image.shape)

# Add subplots
figure, axisarr = plt.subplots(1,2)

# Change from color to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

axisarr[0].imshow(gray_image)
axisarr[1].imshow(gray_image, cmap='gray')
```

```python
# Prints specific grayscale pixel values
# What is the pixel value at x = 400 and y = 300 (on the body of the car)?

x = 400
y = 300

print(gray_image[y,x])
```

```python
# Finds the maximum and minimum grayscale values in this image

max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print('Max: ', max_val)
print('Min: ', min_val)
```

```python
# Create a 5x5 image using just grayscale, numerical values
tiny_image = np.array([[255, 200, 255, 200, 255],
                      [255, 90, 255, 90, 255],
                      [150, 255, 255, 255, 150],
                      [240, 90, 255, 90, 240],
                      [255, 20, 20, 20, 255]])

# To show the pixel grid, use matshow
plt.matshow(tiny_image, cmap='spring')

## TODO: See if you can draw a tiny smiley face or something else!
## You can change the values in the array above to do this
```

```python
# Isolate RGB channels
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

slight_red_image = image.copy()
slight_red_image[:, :, 0] = 100

# Visualize the individual color channels
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15,15))
ax1.set_title('R channel')
ax1.imshow(r, cmap='Reds')
ax2.set_title('G channel')
ax2.imshow(g, cmap='Greens')
ax3.set_title('B channel')
ax3.imshow(b, cmap='Blues')
ax4.set_title('Some red')
ax4.imshow(slight_red_image)

```
