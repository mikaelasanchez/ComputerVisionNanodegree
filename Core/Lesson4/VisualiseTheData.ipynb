# Classifying our Dataset
In order to build a classifier, we need to load in our data and check its distinguishing features
### Import resources

```python
import os
import glob

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random
```

### Load data function

```python
def load_data(dir_name):
    types = glob.glob(os.path.join(dir_name, "*", ""))
    img_list = []
    
    for tp in types:
        imgs = glob.glob(os.path.join(str(tp), "*"))
    
        for img in imgs:
            img_list.append([mpimg.imread(img), tp[len(dir_name):-1]])
            
    return img_list
```

### Visualise some data

```python
img_train_dir = 'data/train/'
image = 0
label = 1

train_data = load_data(img_train_dir)
image1 = train_data[random.randrange(0,119)]
image2 = train_data[random.randrange(120,199)]

_, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.set_title("Label: "+ image1[label] + ", Shape: " + str(image1[image].shape))
ax1.imshow(image1[image])

ax2.set_title("Label: "+ image2[label] + ", Shape: " + str(image2[image].shape))
ax2.imshow(image2[image])
```
